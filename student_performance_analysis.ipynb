```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from matplotlib import pyplot as plt 

plt.style.use("fivethirtyeight")
sns.set_style("white")
```

```python
colors1 = ['#e6ccb2','#b08968','#7f5539'] 
colors2 = ['#ede0d4','#e6ccb2','#ddb892','#b08968','#7f5539','#9c6644','#dda15e']
```

```python
def cat_unmariated_analysis(feature):
  data_ = data[feature].value_counts().sort_values(ascending = True)
  plt.figure(figsize = (20,5),facecolor = 'white')
  plt.subplot(1,2,1)
  plt.barh(y = data_.index,width = data_.values,
           height = 0.7,linewidth = 1,
           color = colors2,edgecolor = 'k')
  

  sns.despine()
  ax = plt.subplot(1,2,2)
  ax.pie(x = data_.values,labels = data_.index,
         colors = colors2,shadow = True,
         radius = 1.2,autopct = "%1.1f%%",
         explode = [0.04 for i in range(data[feature].nunique())])
  ax.add_artist(plt.Circle((0,0),0.04,facecolor = 'white'))
  plt.suptitle(t = feature,x = 0.5,y =1.5,fontstyle = 'italic')
  plt.show()
```

```python
cat_unmariated_analysis('gender')
```

```python
cat_unmariated_analysis('lunch')
```

```python
cat_unmariated_analysis("test preparation course")
```

```python
cat_unmariated_analysis("parental level of education")
```
